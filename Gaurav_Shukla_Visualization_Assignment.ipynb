Que 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
df
```

```python
# Pivot the DataFrame to have Months as columns, Days as rows, and Sales as values
pivot_df = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc=np.sum)
pivot_df
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df, cmap='Set1', annot=True, fmt=".0f", linewidths=0.5)
plt.title('Variation in Sales Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

Que 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
np.random.seed(0)
N = 50
x = np.random.rand(N) * 10  # Random x values
y = np.random.rand(N) * 10  # Random y values
sizes = np.random.rand(N) * 30 + 5  # Random sizes for markers
colors = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50 + 2*x, 30 + 2*y)]  # Random colors for markers
```

```python
plot = figure(title = "My scatter plot", x_axis_label = "X", y_axis_label = "Y")

plot.circle(x, y, size = sizes, color = colors, alpha = 0.6)
show(plot)
```

Que 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook
```

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure
p = figure(x_range=fruits, height=350, title="Counts of Different Fruits")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_label="Counts", line_color="white")

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = max(counts) + 5
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

Que 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
# Creating Bokeh figure
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Addding vertical bars for histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="white")

output_notebook()
show(p)
```

Que 5.  Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
```

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
# Create a new plot
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1),
           tools="hover,pan,wheel_zoom,box_zoom,reset")

# Generate the heatmap
p.image(image=[data_heatmap], x=0, y=0, dh=1, dw=1, palette="Spectral11")

# Add axis labels
p.xaxis.axis_label = 'X'
p.yaxis.axis_label = 'Y'

# Display the plot
show(p)
```
