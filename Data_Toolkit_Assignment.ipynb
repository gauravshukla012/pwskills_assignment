Que. 3 Explain the following terms:

# The difference in np.array, np.asarray and np.asanyarray.

## np.array
Always creates a new array object.

Always copies the input data.
## np.asarray
May return the input array if it is already an ndarray.

Always copies the input data.
## np.asanyarray
May return the input array if it is already an ndarray.

Can convert certain non-array objects to arrays without copying.

# The difference between Deep copy and shallow copy.

## Shallow copy

Creates a new object that contains references to the original object's data.

Changes to the new object's data will also change the original object's data.

## Deep copy

Creates a new object that contains a completely independent copy of the original object's data.

Chages to the new object's data will not affect the original object's data.

Que. 17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
x = np.random.randint(-11, 11, 50)
y = np.random.randint(-11, 11, 50)
```

```python
# here i am defining the funtion to determine the quadrant

def quadrant(x,y):
  if x > 0 and y > 0:
    return 1 # Quadrant I
  elif x < 0 and y > 0:
    return 2 # Quadrant II
  elif x < 0 and y < 0:
    return 3 # Quadrant III
  elif x > 0 and y < 0:
    return 4 # Quadrant IV
  else:
    return 0 # Origin
```

```python
# Now creating my dataframe

df = pd.DataFrame({"x":x, "y":y})
df["quadrant"] = df.apply(lambda row: quadrant(row["x"], row["y"]), axis=1)
df
```

```python
sns.scatterplot(x = "x", y = "y", data=df, hue = "quadrant", legend = "brief")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Quadrant-wise Scatter Plot")
plt.show()
```

Que. 18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'.

```python
from bokeh.io import output_notebook
output_notebook()

from bokeh.plotting import figure, show
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

p = figure(title="sine Wave Function", y_range=(-1.1, 1.1))

p.line(x,y, line_width=2)
p.grid.grid_line_color = None
p.xaxis.axis_label = 'Angle (radians)'
p.yaxis.axis_label = 'Sine'
show(p)
```

Que. 19  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
import random
from bokeh.io import show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
```

```python
# Enable output to the notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(1, 100) for _ in categories]

# Here I am preparing the data
data = {'categories': categories, 'values': values}
source = ColumnDataSource(data=data)

# Creating the figure
p = figure(x_range=categories, height=400, width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Adding the bar renderer
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Adding hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Category"
p.yaxis.axis_label = "Value"

# Display the plot
show(p)
```

Que. 20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'.

```python
import plotly.graph_objects as go
import random
```

```python
x_values = list(range(1, 11))
y_values = [random.randint(1, 100) for _ in x_values]

fig = go.Figure()

fig.add_trace(go.Scatter(x=x_values, y=y_values, mode='lines', name='Random Data'))

fig.update_layout(
    title="Simple Line Plot",
    xaxis_title="X-axis",
    yaxis_title="Y-axis"
)

fig.show()
```
